```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("auto-mpg.csv")
dataframe
```

```python
features = dataframe.drop(["car", "mpg"], axis=1)
labels = dataframe["mpg"]
```

```python
scalar = MinMaxScaler()
```

```python
features = scalar.fit_transform(features)
features
```

```python
standard_scalar = StandardScaler()
```

```python
features = standard_scalar.fit_transform(features)
features
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.20)
```

```python
X = np.append(np.ones((x_train.shape[0], 1)), x_train, axis=1)
weights = np.random.rand(X.shape[1])

print(X[:1])
print(weights)
```

```python
# Cost function
def mean_squared_error(actual, output):
    samples = len(actual)
    cost_sum = 0
    
    for y, y_hat in zip(actual, output):
        cost = (y - y_hat) ** 2
        cost_sum += cost
        
    error = cost_sum / (samples * 2)
    
    return error
```

```python
# Gradient Function #00#0000001
def gradient_step(x, y, y_hat, weights, total_samples, alpha=0.01):

    derivate_error = -(y - y_hat)
#     print(derivate_error)
    derivate_error = x.dot(derivate_error)
#     print(derivate_error)
    weights = weights - alpha/total_samples * derivate_error
    
    return weights
    
```

```python
# Train Function
def train(features, labels, epochs, weights):
    
    errors = []
    
    for e in range(epochs):
#       apply linear regression y_hat = w(i)*x(i) + ... + w(n) * x(n)
        predictions = []
        for data_point in features:
            y_hat = 0
            for w, x in zip(weights, data_point):
                y_hat = y_hat + (w*x)

            predictions.append(y_hat)

#         print(predictions)

#       Calculating Error
        error = mean_squared_error(actual=labels, output=predictions)
        errors.append(error)
#         print(error)
#         print(weights)
#         print("---------------")
#       Applying Gradient Descent   
        for x, y, y_hat in zip(features, labels, predictions):
            weights = gradient_step(x, y, y_hat, weights, len(labels))
        
    return errors, weights
```

```python
errors, final_weights = train(X, y_train, 1000, weights)
```

```python
print(errors)
```

```python
plt.plot(errors)
plt.xlabel("Iteration")
plt.ylabel("Mean Squared Error")
plt.title("Cost function using Gradient Descent")
```

```python
def predict(x_test, weights):
    predictions = []
    for data_point in x_test:
        y_hat = 0
        for w, x in zip(weights, data_point):
            y_hat = y_hat + (w*x)

        predictions.append(y_hat)
        
    return predictions
```

```python
X_test = np.append(np.ones((x_test.shape[0], 1)), x_test, axis=1)
y_pred = predict(X_test, final_weights)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```
